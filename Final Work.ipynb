# Partie PL

```python
import numpy as np  
num_ports = 500
all_weights = np.zeros((num_ports, len(assetschosen1.columns)))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for ind in range(num_ports): 
    # weights 
    weights = np.array(np.random.random(len(assetschosen1.columns))) 
    weights = weights/np.sum(weights)  
	
    # save the weights
    all_weights[ind,:] = weights
	
    # expected return 
    ret_arr[ind] = np.sum((assetschosen1.mean()*weights)*252)

    # expected volatility 
    vol_arr[ind] = np.sqrt(np.dot(weights.T,np.dot(assetschosen1.cov()*252, weights)))

    # Sharpe Ratio 
    sharpe_arr[ind] = ret_arr[ind]/vol_arr[ind]
```

```python
def get_ret_vol_sr(weights):
    weights = np.array(weights)
    ret =np.sum(assetschosen1.mean()*weights)*252
    vol =np.sqrt(np.dot(weights.T,np.dot(assetschosen1.cov()*252,weights)))
    sr = ret/vol 
    return np.array([ret,vol,sr])
```

```python
from scipy.optimize import minimize
```

```python
def neg_sharpe(weights): 
    return get_ret_vol_sr(weights)[2]*-1

neg_sharpe(weights)
```

```python
def check_sum(weights): 
    return np.sum(weights)-1

check_sum(weights)
```

```python
cons=({"type":"eq","fun":check_sum})
```

```python
def allbounds():
    bounds=[]
    for x in range(len(assetschosen1.columns)):
        bounds.append((0,0.1))
    return bounds    
```

```python
def allguess():
    init_guess=[]
    for x in range(len(assetschosen1.columns)):
        init_guess.append(1/len(assetschosen1))
    return init_guess    
```

```python
len(assetschosen1.columns)
```

```python
opt_results=minimize(neg_sharpe,allguess(),method='SLSQP',bounds=allbounds(),constraints=cons)
```
